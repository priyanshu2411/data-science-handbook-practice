```python
mysamples=[]
mylinear=[]
myquardratic=[]
mycubic=[]
myexponential=[]
for i in range(30):
    mysamples.append(i)
    mylinear.append(i)
    myquardratic.append(i**2)
    mycubic.append(i**3)
    myexponential.append(1.5**i)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(mysamples,mylinear,'r-',mysamples,myquardratic,'b-')
```

```python
plt.plot(mysamples,mycubic,'y-',mysamples,myexponential,'g-')
```

```python
plt.figure(figsize=(15,5))
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x,x+0,'-')
plt.plot(x,x+1,'--')
plt.plot(x,x+3,'-.')
plt.plot(x,x+4,':')
plt.plot(x,x+5,'-')
plt.plot(x,x+6,'--')
plt.plot(x,x+7,'-.')
plt.plot(x,x+8,':')
```

```python
plt.plot(x,np.tan(x),':r')
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x,np.cos(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
plt.title('a sine curve')
plt.xlabel("x")
plt.ylabel("f(x)=sin(x)")
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
